```python
# Now we split our data into training vs testing
train_data = df['2016-01-01 00:00':'2016-05-01 00:00']
test_data = df['2016-05-01 00:15':'2016-06-30 23:45']

# Plot the two datasets
plt.clf()
plt.plot(train_data,label="Training data")
plt.plot(test_data,label= "Testing data")
plt.legend()
plt.show()
```

```python
warnings.filterwarnings("ignore") # specify to ignore warning messages

#mod = statsmodels.tsa.arima_model.ARIMA(train_data,order=(1,2,1)).fit(disp=False)
#results = mod.fit()

AIC = []
SARIMAX_model = []
for param in pdq:
        try:
            mod = statsmodels.tsa.arima_model.ARIMA(train_data,order=param)

            results = mod.fit()

            print('ARIMAX{} - AIC:{}'.format(param, results.aic))
            AIC.append(results.aic)
            SARIMAX_model.append([param, param_seasonal])
        except:
            continue
```

```python
print('The smallest AIC is {} for model ARIMAX{}'.format(min(AIC), SARIMAX_model[AIC.index(min(AIC))][0],SARIMAX_model[AIC.index(min(AIC))][1]))
```

```python
# Let's fit this model
mod = sm.tsa.statespace.SARIMAX(train_data,
                                order=SARIMAX_model[AIC.index(min(AIC))][0],
                                seasonal_order=SARIMAX_model[AIC.index(min(AIC))][1],
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results = mod.fit()
```
