```python
import ARIMA_routines as ar
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
```

```python
# The function that comutes the Mean squared error and Root mean squared error

def compute_metric(y_test,y_forecast):
    
    MSE = 0.0
    
    for i in range(0,len(y_test)):
        MSE+= (y_test[i]-y_forecast[i])**2
    
    MSE = MSE/float(len(y_test))
    
    RMSE = np.sqrt(MSE)
    
    return MSE, RMSE
```

```python
# These few cells will fit the ARIMA model to the EURO/USD Data 

# specified to ignore warning messages
warnings.filterwarnings("ignore") 


# We read in the EUR to USD conversion rate
fileName = "data/EURUSD_15m_BID_01.01.2010-31.12.2016.csv"

# Load the data, take only the closing exchange rate data
df = pd.read_csv(fileName, engine='python', skipfooter=3)
df = df[['Time', 'Close']].copy()


# Convert the 'Time' column to appropriate format
df['Time']=pd.to_datetime(df['Time'], format='%Y-%m-%d')
df.set_index(['Time'], inplace=True)


# Now we split our data into training vs testing sets
train_data = df['2016-01-01 00:00':'2016-05-01 00:00']
test_data = df['2016-05-01 00:15':'2016-06-30 23:45']

# This specifies the maximum p,d,q used in the gridsearch
pMax,dMax,qMax = 2,2,2

# We fit the ARIMA model
AIC,RMS, ARIMAX_model, results = ar.fit_ARIMAX(pMax,dMax,qMax,train_data,test_data,"AIC")

# Now that we have fit the model, we generate a forecast based on the fitted model
forecast, stderr, conf = results.forecast(len(test_data))

#-------------------------------------------------------------
# Create a Dataframe with the forecast, create a 'time' column with correct format
df = pd.DataFrame(columns=['DATE', 'FORECAST'])
for i in range(len(test_data)):
    df.loc[i] = [test_data.index[i], forecast[i]]
    
df['DATE']=pd.to_datetime(df['DATE'], format='%Y-%m-%d')
df.set_index(['DATE'], inplace=True)  
#-------------------------------------------------------------

#-------------------------------------------------------------
# Here we create a Dataframe for the upper and lower bounds
# of the forecast with correctly formatted time column
#-------------------------------------------------------------
df_conf_upper = pd.DataFrame(columns=['DATE', 'FORECAST_UPPER'])
df_conf_lower = pd.DataFrame(columns=['DATE', 'FORECAST_LOWER'])

for i in range(len(test_data)):
    df_conf_upper.loc[i] = [test_data.index[i], conf[i][1]]
    df_conf_lower.loc[i] = [test_data.index[i], conf[i][0]]


df_conf_upper['DATE']=pd.to_datetime(df_conf_upper['DATE'], format='%Y-%m-%d')
df_conf_lower['DATE']=pd.to_datetime(df_conf_lower['DATE'], format='%Y-%m-%d')

df_conf_upper.set_index(['DATE'], inplace=True)   
df_conf_lower.set_index(['DATE'], inplace=True)
#-------------------------------------------------------------
```

```python
#============================================
# Compute the metrics of the model
#============================================
df_original = pd.read_csv(fileName, engine='python', skipfooter=3)

# Print out the first few rows of the data, and some summarizing statistics
print(df_original.head(),'\n')
print(df_original.describe(), '\n')

# Convert the Data frame into an array
y_test=[test_data['Close'][k] for k in range(0,len(test_data))]
y_forecast =[df['FORECAST'][k] for k in range(0,len(df))]

# Compute the Mean Square and RMSE value of the model
MSE, RMSE = compute_metric(y_test,y_forecast)
print('MSE: ',MSE,' RMS: ', RMSE)


#============================================
# Here we plot the raw data in the file
#=============================================
plt.plot(train_data,color='blue')
plt.plot(test_data,color='blue')
plt.ylabel('EURO/USD exchange rate',fontsize=20)
plt.xlabel('Date',fontsize=20)
plt.savefig('EURO_USD_exchange_rate.pdf',bboxes='tight')
plt.show()

#============================================
# Here we plot the results of the FIT
#=============================================
plt.clf()
plt.plot(train_data,color='b',label='Training data')
plt.plot(test_data, color='g',label='Testing data')
plt.plot(df_conf_upper,color='r')
plt.plot(df_conf_lower,color='r')
plt.fill_between(df_conf_upper.index,df_conf_lower.iloc[:,0],df_conf_upper.iloc[:,0],alpha=0.2, color='r',label='Prediction')
plt.plot(df,color='r')
plt.legend()
plt.ylabel('EURO/USD exchange rate',fontsize=20)
plt.xlabel('Date',fontsize=20)
plt.savefig('EURO_USD_exchange_rate_with_Forecast.pdf',bboxes='tight')
plt.show()
#=============================================
```

```python
# These few cells will fit the SARIMA model to the airline passenger dataset 

# specified to ignore warning messages
warnings.filterwarnings("ignore") 

# First we read in the sunspot data
fileName = "data/international-airline-passengers.csv"

# Load the data
df = pd.read_csv(fileName, engine='python', skipfooter=3)
# Process month to be a time format
df['Month']=pd.to_datetime(df['Month'], format='%Y-%m-%d')
df.set_index(['Month'], inplace=True)


# Split our data into training vs testing and also validation sets
train_data = df['1949-01-01':'1955-01-01']
test_data = df['1955-02-01':'1956-01-01']
validation_data = df['1956-02-01':'1960-12-01']


# This specifies the maximum p,d,q used in the gridsearch
pMax,dMax,qMax,t = 1,1,2,12


# Perform a grid search, and return the fitting metrics 
AIC,RMS, SARIMAX_model, results = ar.fit_SARIMAX(pMax,dMax,qMax,t,train_data,test_data)

# Show the predictions
pred  =  results.get_forecast(validation_data.index[-1])
pred_ci = pred.conf_int()
pred_mean = pred.predicted_mean

```

```python
#============================================
# Compute the metrics of the model
#============================================

# Print out a statistical description of the dataset
print(df.describe(), '\n')

# Convert the test data and the forecast data to arrays
y_test=[validation_data.iloc[k,0] for k in range(0,len(validation_data))]
y_forecast =[pred_mean[k] for k in range(0,len(pred_mean))]

# Compute the metrics of the model
MSE, RMSE = compute_metric(y_test,y_forecast)
print('MSE: ',MSE,' RMS: ', RMSE)


# Plot the raw data, save to a file
plt.clf()
plt.plot(train_data,color='b')
plt.plot(test_data,color='b')
plt.plot(validation_data,color='b')
plt.ylabel('Number of Airline Passengers',fontsize=15)
plt.xlabel('Date',fontsize=20)
plt.savefig('Airline_Passengers.pdf',bboxes='tight')
plt.show()

# Plot the training data, along with the forecast and confidence intervals
plt.clf()
plt.fill_between(pred_ci.index,pred_ci.iloc[:,0],pred_ci.iloc[:,1],alpha=0.2, color='r')
plt.plot(pred_mean,label="Forecast",color='r')
plt.plot(train_data,label="Training data",color='b')
plt.plot(test_data,color='b')
plt.plot(validation_data,label='Testing data',color='g')
plt.ylabel('Number of Airline Passengers',fontsize=15)
plt.xlabel('Date',fontsize=20)
plt.legend(loc=2)
plt.savefig('Airline_Passengers_Forecast.pdf',bboxes='tight')
plt.show()
```
